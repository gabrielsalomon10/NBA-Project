```python
df = pd.read_csv("common_player_info.csv")
df.head(5)
```

```python
#Expresa el porcentaje de valores nulos de cada columna
missingData = df.isnull().mean() * 100
print(missingData[missingData > 0].sort_values(ascending=False))
```

```python
dfEliminados = df.drop(["first_name", "last_name", "display_last_comma_first", "display_fi_last", "player_slug", "last_affiliation", "dleague_flag", "nba_flag", "games_played_flag", "draft_year", "draft_round", "draft_number", "team_abbreviation", "team_code", "team_city", "playercode", "from_year", "to_year" ], axis=1)
dfEliminados.head(5)
```

```python
#Cambiamos el nombre de la columna
dfEliminados = dfEliminados.rename(columns={'display_first_last': 'player_name'})
dfEliminados.head(5)
```

```python
#Cambiamos el tipo de dato de season_exp
dfEliminados["season_exp"] = dfEliminados["season_exp"].astype(int)
dfEliminados.head(5)
```
