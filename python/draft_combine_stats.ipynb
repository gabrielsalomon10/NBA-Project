```python
#Eliminamos las columnas que no vamos a utilizar
dfEliminados = df.drop(["first_name", "last_name", "height_wo_shoes", "height_wo_shoes_ft_in", "height_w_shoes", "height_w_shoes_ft_in", "wingspan", "wingspan_ft_in", "standing_reach", "standing_reach_ft_in", "body_fat_pct", "hand_length", "hand_width", "modified_lane_agility_time", "bench_press", "spot_fifteen_corner_left", "spot_fifteen_break_left", "spot_fifteen_top_key", "spot_fifteen_break_right", "spot_fifteen_corner_right", "spot_college_corner_left", "spot_college_break_left", "spot_college_top_key", "spot_college_break_right", "spot_college_corner_right", "spot_nba_corner_left", "spot_nba_break_left", "spot_nba_top_key", "spot_nba_break_right", "spot_nba_corner_right", "off_drib_fifteen_break_left", "off_drib_fifteen_top_key", "off_drib_fifteen_break_right", "off_drib_college_break_left", "off_drib_college_top_key", "off_drib_college_break_right", "on_move_fifteen", "on_move_college"], axis=1)
dfEliminados
```

```python
# Identificar columnas con valores nulos
valoresFaltantesColumnas = dfEliminados.columns[dfEliminados.isnull().any()]

# Imprimir las columnas con valores nulos
print("Columnas con valores nulos:\n", valoresFaltantesColumnas)

# Rellenar valores nulos con el promedio de la columna agrupado por paÃ­s
for column in valoresFaltantesColumnas:
    dfEliminados.loc[:, column] = dfEliminados.groupby("player_id")[column].transform(lambda x: x.fillna(0))

# Mostrar la tabla resultante
print("Tabla despuÃ©s de la imputaciÃ³n de valores nulos:\n", dfEliminados)
```
