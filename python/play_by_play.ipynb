```python
import pandas as pd

# Ruta al archivo CSV
path = 'play_by_play.csv'  # Cambia esto por la ruta de tu archivo CSV

# Cargar el archivo CSV en chunks de 100,000 filas
chunkSize = []

# Iterar sobre los chunks
for chunk in pd.read_csv( path, chunksize=100000 ):
    # Aquí puedes procesar cada chunk según sea necesario
    # Por ejemplo, puedes agregarlo a una lista
    chunkSize.append(chunk)

df = pd.concat( chunkSize, ignore_index= True )
df.head(5)
```

```python


# Mostrar la información básica del dataframe
print(df.info())
print("*************************************")

# Mostrar las estadísticas descriptivas del dataframe
print(df.describe())
print("*************************************")

# Visualizar la distribución de una variable numérica
#sns.histplot(df['game_id'])

# Visualizar la relación entre dos variables numéricas
#sns.scatterplot(x='columna_x', y='columna_y', data=df)

# Identificar valores nulos o missing
print("La cantidad de nulos es:\n", df.isnull().sum())

# Identificar outliers en una variable numérica
#Q1 = df['game_id'].quantile(0.25)
#Q3 = df['game_id'].quantile(0.75)
#IQR = Q3 - Q1
#print(df[~((df['game_id'] < (Q1 - 1.5 * IQR)) | (df['game_id'] > (Q3 + 1.5 * IQR)))])
```

```python
missingData = df.isnull().mean() * 100
print(missingData[missingData > 0].sort_values(ascending=False))
```

```python
#Eliminamos las columnas que no vamos a utilizar
dfEliminados = df.drop(["neutraldescription", "pctimestring", "player1_team_id", "player1_team_city", "player1_team_nickname", "player1_team_abbreviation", "score", "person1type", "homedescription", "visitordescription", "scoremargin", "person2type", "person3type", "player2_team_nickname", "player3_team_nickname", "player2_team_abbreviation", "player2_team_id", "player3_team_city", "player2_team_city", "player3_team_abbreviation", "player3_team_id", "wctimestring", "video_available_flag"], axis=1)
```

```python
#Calculo para ver el porcentaje de nulos en cada columna
missingData = dfEliminados.isnull().mean() * 100
print(missingData[missingData > 0].sort_values(ascending=False))
```

```python
# Seleccionar las columnas de interés
columnasEspecificas = [
    'player3_name',  
    'player2_name',
    'player1_name'     
]
```

```python
# Identificar columnas con valores nulos
valoresFaltantesColumnas = dfEliminados.columns[dfEliminados.isnull().any()]

# Imprimir las columnas con valores nulos
print("Columnas con valores nulos:\n", valoresFaltantesColumnas)

# Rellenar valores nulos con el promedio de la columna agrupado por país
for column in valoresFaltantesColumnas:
    dfEliminados.loc[:, column] = dfEliminados.groupby("game_id")[column].transform(lambda x: x.fillna(0))

# Mostrar la tabla resultante
print("Tabla después de la imputación de valores nulos:\n", dfEliminados)
```
